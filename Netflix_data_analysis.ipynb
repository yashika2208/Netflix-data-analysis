```python
import matplotlib.pyplot as plt

df["type"].value_counts().plot(kind="bar")
plt.title("Movies vs TV Shows")
plt.show()
```

Observation
- Movies make up the majority of Netflix's content.
- TV Shows are fewer than Movies.
- Netflix focuses more on movie content.

```python
import matplotlib.pyplot as plt

df["country"].value_counts().head(10).plot(kind="bar", figsize=(10,5))
plt.title("Top 10 Countries Producing Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

Observation:
- The United States has the highest number of Netflix titles.
- India is one of the leading content-producing countries.

```python
df["rating"].value_counts().plot(kind="bar", figsize=(10,5))
plt.title("Content Ratings on Netflix")
plt.xlabel("Rating")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

Observation:
- TV-MA and TV-14 are the most common ratings.
- Netflix primarily offers content for mature audiences.

```python
df["release_year"].value_counts().sort_index().plot(figsize=(12,5))
plt.title("Netflix Content Released by Year")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

Observation:
- The number of titles increased significantly after 2015.
- Content production peaked around 2019–2020.

```python
df['director'].dropna().value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Directors")
plt.xlabel("Director")
plt.ylabel("Number of Titles")
plt.xticks(rotation=45)
plt.show()
```

Observation:
- A few directors have contributed multiple titles.
- Most directors have only a small number of titles on Netflix.

```python
df['listed_in'].str.split(', ').explode().value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Genres")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

Observation:
- International Movies is the most common genre on Netflix.
- Dramas and Comedies are also among the most popular categories.
- Romantic Movies and Children & Family Movies appear less frequently than the top genres.
